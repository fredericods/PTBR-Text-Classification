# Exploratory Data Analysis

## 1) Importing packages and dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

data = pd.read_csv('B2W-Reviews01.csv',sep=';')
data
```

## 2) Features Distributions

#### reviewer_id

```python
display(data.reviewer_id.value_counts())
print('Total Rows:', data.shape[0])
print('Unique Reviewers:', data.reviewer_id.nunique())
```

#### product_id

```python
display(data.product_id.value_counts())
print('Total Rows:', data.shape[0])
print('Unique Products:', data.product_id.nunique())
```

#### site_category_lv1

```python
display(data.site_category_lv1.value_counts())
print('Total Rows:', data.shape[0])
print('Unique Categories:', data.site_category_lv1.nunique())
```

#### reviewer_gender

```python
display(data.reviewer_gender.value_counts())
```

#### target features

```python
freq_overall_rating = (
    pd.DataFrame(data.overall_rating.value_counts()).
    reset_index().
    rename(columns={'index':'rating','overall_rating':'rows'}).
    sort_values(by='rating').
    reset_index(drop=True)
)
freq_overall_rating['Frequency'] = freq_overall_rating.rows/freq_overall_rating.rows.sum()

freq_recommend_to_a_friend = (
    pd.DataFrame(data.recommend_to_a_friend.value_counts()).
    reset_index().
    rename(columns={'index':'recommend_to_a_friend','recommend_to_a_friend':'Rows'}).
    #sort_values(by='Rating').
    reset_index(drop=True)
)
freq_recommend_to_a_friend['Frequency'] = freq_recommend_to_a_friend.Rows/freq_recommend_to_a_friend.Rows.sum()

display(freq_overall_rating)
display(freq_recommend_to_a_friend)
```
